```python
import torch.nn as nn
import torch
import torch.nn.functional as F
from torchsummary import summary
```

```python
class SCNN(nn.Module):
    def __init__(self, num_classes=4):
        super(SCNN, self).__init__()
        self.conv1 = self._make_conv(3, 64, 5, 5, 5)
        self.conv2 = self._make_conv(64, 128, 5, 5, 5)
        self.conv3 = self._make_conv(128, 128, 3, (4,6), (4,6))
        self.drop = nn.Dropout(p=0.5)
        self.fc = nn.Linear(128, num_classes)
        
    def _make_conv(self, in_channel, out_channel, kernel_size, pooling, stride):
        padding = kernel_size / 2
        conv = nn.Sequential(
            nn.Conv2d(in_channel, out_channel, kernel_size, padding=padding, bias=False),
            nn.BatchNorm2d(out_channel),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(pooling, stride=stride)
        )
        return conv
    
    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = x.view(x.size(0), -1)
        x = self.drop(x)
        return self.fc(x)
```

```python
net = SCNN(num_classes=4)
```

```python
def get_images_path(df):
    dst_df = pd.DataFrame(columns=['id', 'age', 'her2', 'p53', 'label'])
    num = 0
    for i in range(len(df)):
        cur_id, age, her2, p53, label = df.iloc[i]
        img_dir = os.path.join('data/train/images', cur_id)
        cur_paths = glob.glob(img_dir + '/*')
        for cur_path in cur_paths:
            if cur_path.split('/')[-1][0] == '.':
                continue
            dst_df.loc[num] = [cur_path, age, her2, p53, label]
            num += 1
    return dst_df
```

```python
train_df = get_images_path(train_df)
```

```python
train_y = train_df['label'].tolist()
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=99)
tmp_df, y = ros.fit_sample(train_df, train_y)
```
